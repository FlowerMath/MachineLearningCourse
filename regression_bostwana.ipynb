```python
import statsmodels
import scipy as sc
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.graphics.regressionplots import plot_leverage_resid2
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("botswana.tsv", sep="\t", index_col=False) 
raw.head()
```

```python
raw.religion.value_counts()
```

```python
raw_drop_nan = raw.dropna()
```

```python
raw_drop_nan.shape
```

```python
raw['nevermarr'] = [0 for i in raw.agefm]
raw['nevermarr'][np.isnan(raw.agefm).values] = 1
```

```python
raw = raw.drop('evermarr', axis=1)
```

```python
raw.head(20)
```

```python
raw.agefm.fillna(value=0, inplace=True)
```

```python
raw.head()
```

```python
raw.heduc[raw.nevermarr == 1] = -1
```

```python
raw.head()
```

```python
raw.heduc.value_counts(dropna=False)
```

```python
raw['idlnchld_noans'] = [0 for i in raw.agefm]
raw['idlnchld_noans'][np.isnan(raw.idlnchld).values] = 1
#raw.idlnchld[raw.idlnchld_noans == 1] = -1
```

```python
raw['heduc_noans'] = [0 for i in raw.agefm]
raw['heduc_noans'][np.isnan(raw.heduc).values] = 1
#raw.heduc[raw.heduc_noans == 1] = -2
```

```python
raw['usemeth_noans'] = [0 for i in raw.agefm]
raw['usemeth_noans'][np.isnan(raw.usemeth).values] = 1
#raw.idlnchld[raw.usemeth_noans == 1] = -1
```

```python
raw.heduc.value_counts(dropna=False)
```

```python
raw.idlnchld.fillna(value=-1, inplace=True)
raw.heduc.fillna(value=-2, inplace=True)
raw.usemeth.fillna(value=-1, inplace=True)
```

```python
raw.head(10)
```

```python
raw.dropna(inplace=True)
```

```python
raw.head(10)
```

```python
raw.shape[0]*raw.shape[1]
```

```python
m1 = smf.ols('ceb ~ age + educ + religion + idlnchld + knowmeth + usemeth + agefm + heduc + urban + electric + radio + tv + bicycle + nevermarr + idlnchld_noans + heduc_noans + usemeth_noans', 
             data=raw)
fitted1 = m1.fit()
#print fitted1.summary()
```

```python
round(fitted1.rsquared,3)
```

```python
sms.het_breushpagan(fitted1.resid, fitted1.model.exog)[1]
```

```python
m2 = smf.ols('ceb ~ age + educ + religion + idlnchld + knowmeth + usemeth + agefm + heduc + urban + electric + radio + tv + bicycle + nevermarr + idlnchld_noans + heduc_noans + usemeth_noans', 
             data=raw)
fitted2 = m2.fit(cov_type='HC1')
#print fitted2.summary()
```

```python
sms.het_breushpagan(fitted2.resid, fitted2.model.exog)[1]
```

```python
m3 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + usemeth + agefm + heduc + urban + electric + bicycle + nevermarr + idlnchld_noans + heduc_noans + usemeth_noans', 
             data=raw)
fitted3 = m3.fit(cov_type='HC1')
#print fitted3.summary()
```

```python
round(m2.fit().compare_f_test(m3.fit())[1], 8)
```

```python
round(m1.fit().compare_f_test(m3.fit())[1], 8)
```

```python
m4 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + agefm + heduc + urban + electric + bicycle + nevermarr + idlnchld_noans + heduc_noans', 
             data=raw)
fitted4 = m4.fit(cov_type='HC1')
print fitted4.summary()
```
